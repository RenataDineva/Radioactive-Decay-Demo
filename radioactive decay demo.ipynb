```python
import numpy as np
import matplotlib.pyplot as plt


N0=1000
lam= 0.1
t=np.linspace(0,50,200)

N=N0*np.exp(-lam*t)

plt.plot(t, N, label="Radioactive Decay", color= "blue")
plt.title("Radioactive Decay Simulation")
plt.xlabel("Time t")
plt.ylabel("Number of Atoms N")
plt.legend()
plt.grid(True)
plt.show()
           
```
